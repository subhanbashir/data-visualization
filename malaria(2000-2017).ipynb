```python
import pandas as pd
import missingno as msno
import plotly.graph_objects as go
```

```python
data1=pd.read_csv('reported_numbers.csv')
```

```python
data1.head()
```

```python
msno.matrix(data1)
```

```python
cases = data1.groupby('Year')['No. of cases'].sum()
deaths = data1.groupby('Year')['No. of deaths'].sum()
```

```python
cases
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=cases.index, 
                         y=cases.values,
                         mode='lines',
                         name='no of cases',
                        ))
```

```python
fig2 = go.Figure()

fig2.add_trace(go.Scatter(x=deaths.index, 
                         y=deaths.values,
                         mode='lines',
                         name='no of deaths',
                        ))
```
